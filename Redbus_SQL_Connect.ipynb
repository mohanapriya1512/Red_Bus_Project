```python
import pymysql
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Cpmbining all the csv files into single df:

# List of CSV file paths

csv_files = ['./bus_details_jk2_csv.csv', './bus_details_kerala1_csv.csv', './bus_details_ktcl1_csv.csv',
            './bus_details_pepsu1_csv.csv','./bus_details_rsrtc1_csv.csv','./bus_details_ts1_csv.csv','./bus_details_hrtc1_csv.csv',
            './bus_details_bihar1_csv.csv','./bus_details_astc1_csv.csv','./bus_details_bengal1_csv.csv'] 

# Read the CSV files into DataFrames
dataframes = [pd.read_csv(file) for file in csv_files]

# Combine the DataFrames into one DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
#Create a table 'redbus_details' :
create_table_sql = """
CREATE TABLE redbus_details (
    id INT NOT NULL AUTO_INCREMENT PRIMARY KEY,
    route_name TEXT,
    route_link TEXT,
    bus_name TEXT,
    bus_type TEXT,
    departing_time TIME,
    duration TEXT,
    arrival_time TIME,
    star_rating FLOAT,
    fare_price DECIMAL(7,2),
    seat_availablity INT
);
"""

print(create_table_sql)
myconnection.cursor().execute(create_table_sql)
myconnection.commit()
```
